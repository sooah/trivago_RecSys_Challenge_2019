```python
import os
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
```

```python
train[0:10]
```

```python
session_t = dict(list(train.groupby(train.session_id)))
```

```python
session_t = train.drop_duplicates(["action_type","reference"], "first")
```

```python
session_t
```

```python
imp = pd.DataFrame(session_t.impressions)
imp = imp[~imp["impressions"].isnull()]
imp = imp.impressions
imp_sp =imp.str.split('|')
imp_sp = imp_sp.apply(lambda x : pd.Series(x))
```

```python
imp_sin = imp_sp.stack().to_frame('imp_sin')
# imp_sin.T
```

```python
pri = pd.DataFrame(session_t.prices)
pri = pri[~pri["prices"].isnull()]
pri = pri.prices
pri_sp = pri.str.split('|')
pri_sp = pri_sp.apply(lambda x : pd.Series(x))
# pri_sp
```

```python
pri_sin = pri_sp.stack().to_frame('pri_sin')
# pri_sin
```

```python
imp_pri = pd.concat([imp_sin, pri_sin], axis = 1)
# imp_pri
```

```python
pos = session_t[session_t["action_type"] == "clickout item"]
# pos
```

```python
pos = pos[["user_id", "step", "action_type", "reference"]]
# pos
```

```python
pos_last = pos.drop_duplicates(["user_id","action_type"], "last")
# pos_last
```

```python
pos_ref = pos_last.reference
pos_ref = pos_ref.tolist()
# pos_ref
```

```python
def pos(data):
    if data in pos_ref:
        return 1
    else:
        return 0
```

```python
imp_pri["positive"] = imp_pri['imp_sin'].apply(pos)
```

```python
imp_pri
```

```python
imp_pri.to_csv(path_or_buf = os.path.abspath('imp_pri.csv'))
```
