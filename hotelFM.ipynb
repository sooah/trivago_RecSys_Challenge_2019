```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df1_path = "train.csv"
#df2_path = "test.csv"
```

```python
df1 = pd.read_csv(df1_path)
#df2 = pd.read_csv(df2_path)
#df = df1.append(df2)
df =df1
```

```python
print(df.shape)
df.head()
```

```python
def make_encoding_label_dict(col_unique):
    encoded_dict = {}
    for idx, unique in enumerate(col_unique):
        encoded_dict[unique] = idx + 1

    return encoded_dict


# get new encoding result
def get_newcode(key, label_dict):
    com_len = len(label_dict)
    if key == np.nan:
        return
    else:
        if key in label_dict:
            return label_dict[key]
        else:
            return com_len + 1
```

```python
df = df.fillna(value = '0')
```

```python
encoded_action_dict = make_encoding_label_dict(df.action_type.value_counts().index.tolist())
```

```python
encoded_st_dict = make_encoding_label_dict(df.step.value_counts().index.tolist())
```

```python
encoded_session_dict = make_encoding_label_dict(df.session_id.value_counts().index.tolist())
```

```python
encoded_city_dict = make_encoding_label_dict(df.city.value_counts().index.tolist())
```

```python
df = df[['session_id', 'step', 'action_type', 'reference', 'city']]
```

```python
encoded_ref_dict = make_encoding_label_dict(df.reference.value_counts().index.tolist())
```

```python
encoded_ref_dict
```

# Reference is currently represented by dictionary ( Only step is unchanged)

```python
df['step'] = df['step'].astype('str')
df['session_id'] = df['session_id'].apply(lambda x: get_newcode(x, encoded_session_dict))
df['action_type'] = df['action_type'].apply(lambda x: get_newcode(x, encoded_action_dict))
df['reference'] = df['reference'].apply(lambda x: get_newcode(x, encoded_ref_dict))
df['city'] = df['city'].apply(lambda x: get_newcode(x, encoded_city_dict))
```

```python
print(df.shape)
df.head()
```

```python
col_len_dict = {'session_id': len(encoded_session_dict), 'step': len(encoded_st_dict), 'action_type': len(encoded_action_dict), 'reference': len(encoded_ref_dict), 'city': len(encoded_city_dict)}
col_accum_index_dict = {}
cumulative = 0
for key, value in col_len_dict.items():
    col_accum_index_dict[key] = cumulative
    cumulative = cumulative + value

train_df, test_df = train_test_split(df, test_size=0.2)
```

```python
txt_file = open('trainTXT.txt', 'w')
for idx, row in train_df.iterrows():
    vec = []
    label = row['action_type']
    vec.append(str(label))
    row = row.drop(labels=['action_type'])
    for key, value in row.items():
        if col_len_dict[key] == 1:
            if value != '0':
                col_idx = col_accum_index_dict[key]
                out_val = value
                vec.append(str(col_idx) + ":" + str(out_val))
        else:
            col_idx = col_accum_index_dict[key] + (int(value) - 1)
            out_val = 1
            vec.append(str(col_idx) + ":" + str(out_val))
    txt_file.write("%s\n" % " ".join(vec))
txt_file.close()
```

```python
txt_file = open('testTXT.txt', 'w')
for idx, row in test_df.iterrows():
    vec = []
    label = row['action_type']
    vec.append(str(label))
    row = row.drop(labels=['action_type'])
    for key, value in row.items():
        if col_len_dict[key] == 1:
            if value != '0':
                col_idx = col_accum_index_dict[key]
                out_val = value
                vec.append(str(col_idx) + ":" + str(out_val))
        else:
            col_idx = col_accum_index_dict[key] + (int(value) - 1)
            out_val = 1
            vec.append(str(col_idx) + ":" + str(out_val))
    txt_file.write("%s\n" % " ".join(vec))
txt_file.close()
```
