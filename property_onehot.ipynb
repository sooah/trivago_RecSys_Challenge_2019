```python
import os
import numpy as np
import pandas as pd
```

```python
meta = pd.read_csv('item_metadata.csv')
meta.head()
```

```python
pro = pd.DataFrame(meta.properties)
split = pro.properties.str.split('|')
split_ = split.apply(lambda x : pd.Series(x))
split_sin = split_.stack().reset_index(level = 1, drop = True).to_frame('property_single')
```

```python
split_sin
```

```python
property_single = split_sin['property_single'].unique()
pro_one_hot_encoded = pd.get_dummies(split_sin.property_single)
```

```python
item = pd.DataFrame(meta.item_id)
```

```python
pro_ = pd.concat([item, pro_one_hot_encoded], axis = 1)
```

```python
pro_
```

```python
pro_onehot_sum = pro_.groupby(pro_.item_id).sum()
```

```python
pro_.to_csv(path_or_buf = os.path.abspath('property_onehot_each.csv'))
```

```python
pro_onehot_sum.to_csv(path_or_buf = os.path.abspath('property_onehot_sum.csv'))
```
