```python
import pandas as pd
import numpy as np
```

```python
md = pd.read_csv("train.csv")
```

```python
ac = pd.DataFrame(md.action_type)
poi = md[ac['action_type'] == "search for poi"]
des = md[ac['action_type'] == "search for destination"]
d = pd.concat([poi, des])
```

```python
city = d["city"]
```

```python
city_sp = city.str.split(',')
```

```python
city_sp_ = city_sp.apply(lambda x : pd.Series(x))
```

```python
city_sp_.columns = ['city', 'country', 'nan']
```

```python
city_ = city_sp_.drop(["nan"], axis = 1)
```

```python
city_
```

```python
city_country = city_["country"]
```

```python
city_country
```

```python
ref = d["reference"]
```

```python
ref_sp = ref.str.split(',')
```

```python
ref_sp_ = ref_sp.apply(lambda x : pd.Series(x))
```

```python
ref_sp_.columns = ['landmark', 'country', 'nan']
```

```python
ref_ = ref_sp_.drop(["nan"], axis = 1)
```

```python
ref_region = ref_[~ref_["country"].isnull()]
```

```python
ref_country = ref_region["country"]
```

```python
region = pd.concat([city_country, ref_region],axis = 1)
region.columns = ["city_country", "ref_landmark","ref_country"]
```

```python
region
```

```python
region = region[~region["ref_country"].isnull()]
```
