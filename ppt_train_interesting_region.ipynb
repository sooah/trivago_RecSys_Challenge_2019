```python
import pandas as pd
import numpy as np
```

```python
md = pd.read_csv("train.csv")
```

```python
ac = pd.DataFrame(md.action_type)
poi = md[ac['action_type'] == "search for poi"]
des = md[ac['action_type'] == "search for destination"]
d = pd.concat([poi, des])
```

```python
city = d["city"]
```

```python
city_sp = city.str.split(',')
```

```python
city_sp_ = city_sp.apply(lambda x : pd.Series(x))
```

```python
city_sp_.columns = ['city', 'country', 'nan']
```

```python
city_ = city_sp_.drop(["nan"], axis = 1)
```

```python
city_
```

```python
city_country = city_["country"]
```

```python
city_country
```

```python
ref = d["reference"]
```

```python
ref_sp = ref.str.split(',')
```

```python
ref_sp_ = ref_sp.apply(lambda x : pd.Series(x))
```

```python
ref_sp_.columns = ['landmark', 'country', 'nan']
```

```python
ref_ = ref_sp_.drop(["nan"], axis = 1)
```

```python
ref_region = ref_[~ref_["country"].isnull()]
```

```python
ref_region
```

```python
ref_country = ref_region["country"]
```

```python
region = pd.concat([city_country, ref_region],axis = 1)
region.columns = ["city_country", "ref_landmark","ref_country"]
```

```python
region
```

```python
region = region[~region["ref_country"].isnull()]
```

```python
region
```

```python
region_eq = region[region["city_country"] == region["ref_country"]]
```

```python
region_eq.shape
```

```python
reg_uniq, reg_c = np.unique(region_eq["ref_landmark"], return_counts = True)
```

```python
reg_uniq = np.reshape(reg_uniq, [reg_uniq.shape[0],1])
reg_c = np.reshape(reg_c, [reg_c.shape[0],1])
reg_ = np.concatenate((reg_uniq, reg_c), axis = 1)
reg_
```

```python
reg = pd.DataFrame(reg_, columns = ["reg_uniq", "reg_c"])
```

```python
reg = reg.sort_values(by = ["reg_c"], ascending = False)
reg__ = reg[reg["reg_c"]>1]
```
