```python
city_sp = city.str.split(',')
city_sp_ = city_sp.apply(lambda x : pd.Series(x))
city_sp_.columns = ['city', 'country', 'nan']
city_ = city_sp_.drop(["nan"], axis = 1)
```

```python
kr_d_imp = pd.DataFrame(train_kr_d.impressions)
kr_d_imp.columns = ["imp"]
kr_d_imp = kr_d_imp[~kr_d_imp["imp"].isnull()]
kr_d_imp = kr_d_imp.imp
```

```python
kr_d_imp_sp = kr_d_imp.str.split('|')
kr_d_imp_sp_ = kr_d_imp_sp.apply(lambda x : pd.Series(x))
```

```python
kr_d_imp_sp_ 
```

```python
kr_d_imp_sp_ = kr_d_imp_sp_.astype(np.float)
kr_d_imp_sp_ = np.asarray(kr_d_imp_sp_).flatten()
kr_d_imp_sp_ = np.reshape(kr_d_imp_sp_, [kr_d_imp_sp_.shape[0],1])
kr_d_imp_sp_ = kr_d_imp_sp_[~np.isnan(kr_d_imp_sp_)]
```

```python
[imp_uniq, imp_c] = np.unique(kr_d_imp_sp_, return_counts = True)
imp_uniq = np.reshape(imp_uniq, (imp_uniq.shape[0],1))
imp_c = np.reshape(imp_c, (imp_c.shape[0],1))
imp_ = np.concatenate((imp_uniq, imp_c), axis = 1)
kr_imp = pd.DataFrame(imp_, columns = ["imp_uniq", "imp_c"])
kr_imp = kr_imp.sort_values(by = ["imp_c"], ascending = False)
kr_imp = kr_imp[~kr_imp["imp_uniq"].isnull()]
```

```python
kr_imp.T
```

```python
imp_list = pd.DataFrame(kr_imp.imp_uniq)
imp_list
```

```python
meta = pd.read_csv("item_metadata.csv")
item = pd.DataFrame(meta.item_id)
imp_exist = imp_list[imp_list["imp_uniq"].isin(item.item_id)]
```

```python
pro_imp = meta[meta["item_id"].isin(imp_exist.imp_uniq)]
pro_imp_split = pro_imp.properties.str.split('|')
pro_imp_split_ = pro_imp_split.apply(lambda x : pd.Series(x))
pro_imp_split_.head()
```

```python
pro_imp_split_sin = pro_imp_split_.stack().reset_index(level=1,drop=True).to_frame('property_single')
property_single = pro_imp_split_sin['property_single'].unique()
property_encoded = pd.get_dummies(pro_imp_split_sin.property_single)
property_sum = property_encoded.sum(axis = 0)
property_sum = property_sum.sort_values(ascending = False).to_frame().T
```
